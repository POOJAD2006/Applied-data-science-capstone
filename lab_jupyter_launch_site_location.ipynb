```python
# Task 2: Mark the success/failed launches for each site on the map
# Initialize the map centered at NASA JSC
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Loop through each launch record in the dataset
for idx, row in spacex_df.iterrows():
    # Choose marker color based on success/failure
    if row['class'] == 1:
        marker_color = 'green'  # success
    else:
        marker_color = 'red'    # failure
    
    # Add a circle marker
    folium.Circle(
        location=[row['Lat'], row['Long']],
        radius=2000,
        color=marker_color,
        fill=True,
        fill_color=marker_color,
        fill_opacity=0.6
    ).add_child(folium.Popup(f"{row['Launch Site']} | Class: {row['class']}")).add_to(site_map)
    
    # Add a text label
    folium.map.Marker(
        [row['Lat'], row['Long']],
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 10; color:#000000;"><b>%s</b></div>' % row['Launch Site'],
        )
    ).add_to(site_map)

# Display the map
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
# TASK 3: Calculate the distances between a launch site and its proximities

# Helper distance calculation using Haversine Formula
from math import sin, cos, sqrt, atan2, radians
import folium

def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km
    dlat = radians(lat2 - lat1)
    dlon = radians(lat2 - lon1)
    alat1 = radians(lat1)
    alat2 = radians(lat2)

    a = sin(dlat / 2)**2 + cos(alat1) * cos(alat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Pick launch site (CCAFS LC-40)
launch_site = [28.562302, -80.577356]

# Pick a nearby coastline point from dataset
coastline = [28.563197, -80.568347]

# Calculate distance
distance = calculate_distance(
    launch_site[0], launch_site[1], coastline[0], coastline[1]
)

# Draw a PolyLine to represent the distance visually
distance_text = f"{distance:.2f} km"
line = folium.PolyLine(
    locations=[launch_site, coastline],
    color='blue',
    weight=2,
    popup=distance_text
)

# Add launch site marker and coastline marker with popup
folium.Marker(location=launch_site, popup='Launch Site: CCAFS LC-40').add_to(site_map)
folium.Marker(location=coastline, popup='Coastline Point').add_to(site_map)

# Add the line showing distance
site_map.add_child(line)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
